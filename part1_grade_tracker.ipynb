```python
# Task 1 - Data Parsing & Profile Cleaning
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

```python
# task 1.1 - Loop through raw_students and for each student, produce a cleaned version.

# declaring empty list
cleaned_students=[]

# Loop through each raw student record
for student in raw_students:
  name=student['name'].strip().title() # removes extra spaces and convert to Title Case

  roll=int(student['roll']) # Converts roll number from string to integer

  marks=[int(m) for m in student['marks_str'].split(',')] # splits marks_str and converts the strings into a list of integers

  # created dict of cleaned student records
  cleaned={
      'name':name,
      'roll':roll,
      'marks':marks
  }

  # Add cleaned student to the new list
  cleaned_students.append(cleaned)

print(cleaned_students)
```

```python
# task 1.2 - For each cleaned student, verify the name is valid: check that every word in the name contains only alphabetic characters.

# loop through the entire list of dictionaries
for students in cleaned_students:

  # in each dict it splits the name value into list and checks if all contains only alphabets
  if all(word.isalpha() for word in students['name'].split()):
    print(students['name']+" ✓ Valid name")
  else:
    print(students['name']+" ✗ Invalid name")

```

```python
# task 1.3 - Print a formatted profile card for each cleaned student using f-strings
for students in cleaned_students:
  print(f'{"="*40}\nStudent : {students['name']}\nRoll No : {students['roll']}\nMarks: {students['marks']}\n')
```

```python
# task 1.4 - After processing all students, print the name in ALL CAPS and lowercase for the student with roll number 103.
for student in cleaned_students:
  if student['roll']==103:
    print(student['name'].upper())
    print(student['name'].lower())
```

```python
# Task 2 - Marks Analysis Using Loops & Conditionals
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
```

```python
# task 2.1 - Using a for loop, print each subject alongside its marks and a grade label based on this scheme:
for i in range(len(subjects)):
  if marks[i]>=90:
    print(f'{subjects[i]} {marks[i]} A+')
  elif marks[i]>=80:
    print(f'{subjects[i]} {marks[i]} A')
  elif marks[i]>=70:
    print(f'{subjects[i]} {marks[i]} B')
  elif marks[i]>=60:
    print(f'{subjects[i]} {marks[i]} C')
  else:
    print(f'{subjects[i]} {marks[i]} F')
```

```python
# task 2.2 - Calculate and print: Total marks, Average marks, Highest scoring subject, Lowest scoring subject

# Total marks (sum of all subjects)
Total_marks=0
for i in marks:
  Total_marks=Total_marks+i
print(Total_marks)

# Average marks (rounded to 2 decimal places)
average_marks=round(Total_marks/len(marks),2)
print(f'{average_marks:.2f}')

# Highest scoring subject (name + marks)
highest=marks[0]
for i in range(len(marks)-1):
  if marks[i+1]>highest:
    highest=marks[i+1]
print(highest)

# Lowest scoring subject (name + marks)
lowest=marks[0]
for i in range(len(marks)-1):
  if marks[i+1]<lowest:
    lowest=marks[i+1]
print(lowest)
```

```python
# task 2.3 - Using a while loop, simulate a marks-entry system that allows adding new subjects
subject_list=[]
mark_list=[]
sub_flag=0

subject=str(input('enter the subject: '))
#enter the string and check if it is string
if  (subject.lower()=='done'):
  sub_flag=1  # converts to lower case and sets flag to 1 to check when user enters done

while(sub_flag!=1):
  # while user is not done, initialize valid to be false
  valid=False
  while(valid==False):
    # assuming invalid number entry, keep looping
    try:
      mark=int(input(f'enter the marks for the {subject}:  '))
      if mark<0 or mark>100:
        # check if number not between 0-100, if yes continue the valid loop else exit and move to next subject
        print("Please enter marks between 0 and 100.")
        continue
      break
    except ValueError:
      # if any error occurs continue valid loop
      print("\nPlease enter a valid number.")
  # if number is valid add it to respective new lists
  subject_list.append(subject)
  mark_list.append(mark)
  print('\n')
  subject=input('enter the subject: ')
  # continue checking subject till user press done
  if  (subject.lower()=='done'):
    sub_flag=1

updated_marks=marks+mark_list
print(f'\nNumber of new subjects added: {len(subject_list)}')

print(f'updated average: {round(sum(updated_marks)/len(updated_marks),2)}')
```

```python
# Task 3 - Class Performance Summary
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

# initialize pass_students_count, fail_students_count, topper_average, topper_name, class_average
pass_students_count=0
fail_students_count=0
topper_average=sum(class_data[0][1])/len(class_data[0][1])
topper_name=class_data[0][0]
class_average=0
print(f'{'Name':17} | {'Average':7} | Status')
print("-------------------------------------")
for student in class_data:
  average_status='Fail'
  # initially assume particular student average_status as fail and calculate its average
  student_average=round(sum(student[1])/len(student[1]),2)
  # add student_average for class average calculation
  class_average=class_average+student_average
  # check is particular student average is greater than currently stored topper_average
  if student_average>topper_average:
    # if yes, update topper_average and store student name
    topper_average=student_average
    topper_name=student[0]
  # check if particular student average greater than 60, if yes change average_status to pass else fail, also log pass/fail count
  if student_average>=60:
    average_status='Pass'
    pass_students_count+=1
  else:
    average_status='Fail'
    fail_students_count+=1
  print(f'{student[0]:17} | {student_average:7.2f} | {average_status}')
# Number of students who passed and who failed
print(f'\nNumber of students passed: {pass_students_count}')
print(f'Number of students failed: {fail_students_count}')
# Class topper (name + average)
print(f'\nClass topper: {topper_name} {topper_average:.2f}')
# Class average (average of all five students' averages)
print(f'\nClass average: {class_average/len(class_data):.2f}')
```

```python
# Task 4 - String Manipulation Utility
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# Strip leading and trailing whitespace from essay. Store this as clean_essay and use it for all steps below.
clean_essay=essay.strip()
print("Clean Essay:")
print(clean_essay)

# Convert clean_essay to Title Case and print it.
clean_essay_title=clean_essay.title()
print("\nTitle Case:")
print(clean_essay_title)

# Count how many times the word "python" appears in clean_essay (case-insensitive). Print the count.
python_count=clean_essay.count("python")
print("\nCount of 'python':")
print(python_count)

# Replace every occurrence of "python" in clean_essay with "Python 🐍" and print the result.
clean_essay_replace=clean_essay.replace("python","Python 🐍")
print("\nReplaced Essay:")
print(clean_essay_replace)

# Split clean_essay into individual sentences by splitting on ". " (a period followed by a space). Print the resulting list.
clean_essay_split=clean_essay.split(". ")
print("\nSentence list:")
print(clean_essay_split)

# Print each sentence on a new line, numbered starting from 1. Add a "." at the end of each sentence if it does not already end with one.
print("\nNumbered Sentences:")
for sentence in range(len(clean_essay_split)):
  new_line_sentence=clean_essay_split[sentence]
  if not new_line_sentence.endswith("."):
        new_line_sentence = new_line_sentence + "."
  print(f'{sentence+1}. {new_line_sentence}')
```
